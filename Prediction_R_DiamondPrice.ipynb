```r
# Create two histograms of the price variable
# and place them side by side on one output image.
# The first plot should be a histogram of price
# and the second plot should transform
# the price variable using log10.

# Set appropriate bin widths for each plot.
# ggtitle() will add a title to each histogram.

library(gridExtra)

## SCATTER PLOT:
# p1=ggplot(aes(x=carat, y=price), data=diamonds)+
#     geom_point()+
#     ggtitle("price")

# p2=ggplot(aes(x=carat, y=price), data=diamonds)+
#     geom_point()+
#     scale_y_log10() +
#     ggtitle("log10(price)")

## HISTOGRAM

p1=ggplot(aes(x=price), data=diamonds)+
    geom_histogram(binwidth=100, color='white', fill='orange')+
    ggtitle("Price")

p2=ggplot(aes(x=price), data=subset(diamonds, !is.na(price)))+
    geom_histogram(binwidth=0.01, color='white', fill='blue')+
## we have kept the binwidth too small because, this is log scale
    scale_x_log10() +
    ggtitle("log10(Price)")
grid.arrange(p1, p2, ncol=1)

## Now, why two peaks in log scale and what does it say about the demand?
## Using logscale makes the distributions roughly normal with two modes
## that says there are two local maxima representing two customer segments,
## low paying and high paying
```

```r
## Scatter plot transformation
ggplot(aes(x=carat, y=price), data=diamonds) + 
    geom_point() +
    scale_y_log10() +
    ggtitle("log10(price) vs carat")
## We used log scale for price, bcoz it has a very high range
## Lets try to improve further, 
## use cube root of carat 
```

```r
## create a new function to transform the carat variable
cuberoot_trans = function() trans_new('cuberoot',
                                     transform=function(x) x^(1/3),
                                     inverse=function(x) x^3)
## cuberoot_trans funct takes the cuberoot of any input variable
## it also has an inverse function to undo that operation which
## we need to display the plot correctly
                                         
## trans_new() is in scales package.
ggplot(aes(x=carat, y=price), data=diamonds) + 
    geom_point() +
    scale_x_continuous(trans=cuberoot_trans(), limits=c(0.2,3), 
                       breaks=c(0.2,0.5,1,2,3)) +
    scale_y_continuous(trans=log10_trans(), limits=c(350, 15000), 
                       breaks=c(350,1000,5000,10000,15000)) +
    ggtitle("log10(price) vs cube root carat")                                         

## Look at below, the plot looks linear and we can use linear model
```

```r
# Add a layer to adjust the features of the scatterplot. Set the 
# transparency to one half,the size to three-fourths,and jitter the points.


ggplot(aes(carat, price), data = diamonds) + 
  geom_point(alpha=1/2,size=3/4,position="jitter",color='red') + 
  scale_x_continuous(trans = cuberoot_trans(), limits = c(0.2, 3),
                     breaks = c(0.2, 0.5, 1, 2, 3)) + 
  scale_y_continuous(trans = log10_trans(), limits = c(350, 15000),
                     breaks = c(350, 1000, 5000, 10000, 15000)) +
  ggtitle('Price (log10) by Cube-Root of Carat')
```

```r
## Colors for qualitative data(factors/categories):

## We saw that there is almost a linear relationship between log10
## of price and cube root of carat
## However price also depends on other features eg..clarity

## color the points by clarity.
#?diamonds
# I1 (worst), SI1, SI2, VS1, VS2, VVS1, VVS2, IF (best)
library(RColorBrewer)

ggplot(aes(carat, price), data = diamonds) + 
  geom_point(aes(color=clarity),alpha=1/2,size=3/4,position="jitter") + 
  scale_color_brewer(type = 'div', guide = guide_legend(title='Clarity',
                    reverse=T, override.aes=list(alpha=1,size=2))) +
  scale_x_continuous(trans = cuberoot_trans(), limits = c(0.2, 3),
                     breaks = c(0.2, 0.5, 1, 2, 3)) + 
  scale_y_continuous(trans = log10_trans(), limits = c(350, 15000),
                     breaks = c(350, 1000, 5000, 10000, 15000)) +
  ggtitle('Price (log10) vs\n Cube-Root of Carat and Clarity')

## We could also have done, by adding aes(color=clarity) in ggplot()
## in place of geom_point()

## We can see high clarity corresponds to the top cluster of points 
## and that implies higher price.
```

```r
## color the points by cut

ggplot(aes(carat, price), data = diamonds) + 
  geom_point(aes(color=cut),alpha=1/2,size=3/4,position="jitter") + 
  scale_color_brewer(type = 'div', guide = guide_legend(title='Cut',
                    reverse=T, override.aes=list(alpha=1,size=2))) +
  scale_x_continuous(trans = cuberoot_trans(), limits = c(0.2, 3),
                     breaks = c(0.2, 0.5, 1, 2, 3)) + 
  scale_y_continuous(trans = log10_trans(), limits = c(350, 15000),
                     breaks = c(350, 1000, 5000, 10000, 15000)) +
  ggtitle('Price (log10) vs\n Cube-Root of Carat and Cut')

## We can see that for the same carat, price increases as the 
## quality of cut improves. Further, most of the diamonds have 
## ideal or premium cut.
```

```r
## color the points by color
# ?diamonds
# colour, from J (worst) to D (best)
## Here order is opposite, so remove reverse=T from scale_color_brewer

ggplot(aes(carat, price), data = diamonds) + 
  geom_point(aes(color=color),alpha=1/2,size=3/4,position="jitter") + 
  scale_color_brewer(type = 'div', guide = guide_legend(title='Color',
                    override.aes=list(alpha=1,size=2))) +
  scale_x_continuous(trans = cuberoot_trans(), limits = c(0.2, 3),
                     breaks = c(0.2, 0.5, 1, 2, 3)) + 
  scale_y_continuous(trans = log10_trans(), limits = c(350, 15000),
                     breaks = c(350, 1000, 5000, 10000, 15000)) +
  ggtitle('Price (log10) vs\n Cube-Root of Carat and Color')

## We can see that for the same carat, price increases as the 
## color improves. 
```

```r
## Linear regression model in R
## I() :  means as it is; it tells R to use the expression
## inside the I() wrapper to transform the variable before using
## it in the regression

m1 = lm(I(log(price)) ~ I(carat^(1/3)), data = diamonds)
m2 = update(m1, ~. + carat)
m3 = update(m2, ~. + cut)
m4 = update(m3, ~. + color)
m5 = update(m4, ~. + clarity)

mtable(m1, m2, m3, m4, m5, sdigits = 3) #for 3 significant digits
# summary(m1) # Will give summary
```

```r
## Create a column, logprice
## select only the diamonds certified by GIA and cheaper 
## than 10000
diamondsbig$logprice = log(diamondsbig$price)

m1 = lm(logprice ~ I(carat^(1/3)), 
        data=diamondsbig[diamondsbig$price < 10000 
                         & diamondsbig$cert == 'GIA', ])

m2 = update(m1, ~. + carat)
m3 = update(m2, ~. + cut)
m4 = update(m3, ~. + color)
m5 = update(m4, ~. + clarity)

mtable(m1, m2, m3, m4, m5, sdigits = 3) #for 3 significant digits
# summary(m1) # Will give summary
```

```r
## Predict the price for a diamond:

thisDiamond = data.frame(carat=1.00, cut='V.Good',
                        color='I',clarity='VS1')
modelEstimate = predict(m5, newdata = thisDiamond,
                       interval ='prediction', level=0.95)
exp(modelEstimate)
```
